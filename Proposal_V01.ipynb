```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
```

# Getting Fundamentals from Yahoo Finance
**Can be Automated for any list of securities**

### Here, Apple, Tesla, Walmart and Gold

```python
assets=['AAPL','TSLA','WMT','GC=F']
dict_dfs=({'dfs': [],
          })
##
print('Getting Fundamentals of:\n')
for sec in assets:
    print(sec)
    #,end="\r")

    url = 'https://finance.yahoo.com/quote/'+sec+'/key-statistics?p='+sec
    r  = requests.get(url)
    data = r.text
    #
    soup = BeautifulSoup(data,features="html.parser")
    ##
    rows=soup.find_all('td')
    #
    parameters=[]
    val=[]
    for row in rows[::2]:
        parameters.append(row.get_text(',').split(',')[0])
    #
    for row in rows[1::2]:
        val.append(row.get_text(',').split(',')[0])
    ##
    df=pd.DataFrame(index=parameters,columns=[sec])
    df[sec]=val
    df.index.name='Asset'
    df=df.transpose()
    #
    n=np.random.randint(2,15)
    print('Sleeping for:',n)
    time.sleep(n)
    #print(df)
    dict_dfs['dfs'].append(df)

##
#df_fundamentals=pd.DataFrame()
df_fundamentals=dict_dfs['dfs'][0]
for i in range(1,len(dict_dfs['dfs'])):
    try:
        df_fundamentals=pd.concat([df_fundamentals,
            dict_dfs['dfs'][i]],
            join='outer',axis=0)
    except:
        pass
```
